```
from sklearn.ensemble import RandomForestClassifier
from sklearn import datasets, tree, metrics
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold, cross_val_score, train_test_split, StratifiedKFold, LeaveOneOut
import numpy as np
from sklearn.pipeline import Pipeline
```

```
wisconsin = datasets.load_breast_cancer()
print(wisconsin.keys())
X = wisconsin.data
y = wisconsin.target
```

```
seed = 421
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=seed, stratify=y)
```

```
rf = RandomForestClassifier()
bal_accs = []
feat_imps = []

for i in range(100):
    rf.fit(X_train, y_train)
    y_pred = rf.predict(X_test)
    bal_acc = metrics.balanced_accuracy_score(y_test, y_pred)
    bal_accs.append(bal_acc)

    feat_imps.append(rf.feature_importances_)

feat_imps = np.array(feat_imps)

for i in range(30):
    plt.title("Feature importance of {}".format(wisconsin["feature_names"][i]))
    plt.hist(feat_imps[:,i])
    plt.show()

print(bal_accs)
```
